Хочу написать функцию `pretty_print`, которая принимает на вход одну короткую строку. Если строка длиннее 20 символов, то нужно напечатать "Слишком длинная строка". Если строка короче или равна 20 символам, то нужно сделать чтобы строка выгдядела вот так:

```
#########
# anton #
#########
```

````python
"#" * 20
````

````python
def pretty_print(string):
    if len(string) > 20:
        print("Слишком длинная строка")
    else:
        n_hashes = len(string) + 4
        print("#" * n_hashes)
        print("# " + string + " #")
        print("#" * n_hashes)

pretty_print("sadfgh")
pretty_print("fdsgsfdgvhhdsfghdsfdggh")
print("всё отлично!")
````

````python
values = ["abcd", "bda", "aab", "0xabadbabe", "0xdeadbeef"]

sorted(values, key=len)
````

Как работает key?

1. Применяет функцию, которую ему подали к каждому элементу списка `["abcd" -> 4, "aab" -> 3, "bda" -> 3, "0xabadbabe" -> 10, "0xdeadbeef" -> 10]`

2. `["aab" -> 3, "bda" -> 3, "abcd" -> 4, "0xabadbabe" -> 10, "0xdeadbeef" -> 10]`

3. `["aab", "bda", "abcd", "0xabadbabe", "0xdeadbeef"]`

## кастомная функция для ключа

````python
def sort_by_a(string):
    return string.count('a')

values = ["abcd", "aab", "bda", "0xabadbabe", "0xdeadbeef"]

print(sorted(values, key=sort_by_a))

print(sorted(values, key=lambda x: x.count('a')))
````

````python
new_values = [[3, 'abt'], [2, 'sdfsd'], [5, 'bcg'], [1, '0rg'], [2, 'ab']]

def get_second(elements):
    return elements[1]

print(sorted(new_values, key=get_second))
print(sorted(new_values, key=lambda elements: elements[1]))
````

Задача: отсортировать список по второму элементу, а если вторые элементы одинаковые -- отсортировать по первым. То есть нужно сортировать как бы по объектам задом наперёд.

````python
new_values = [[3, 'abt'], [2, 'sdfsd'], [5, 'ab'], [1, '0rg'], [2, 'ab']]

def backwardsort(elements):
    return [elements[1], elements[0]]

print(sorted(new_values, key=backwardsort))
````

## Рекурсия


Рекурсия -- это про повторение.

Чем рекурсия отличается от цикла?


Рекурсия &ndash; применение некоторого правила к результату примения этого же самого правила.


$f(f(x))$ &ndash; рекурсия.

````python
def dummy_sum(n):
    if n == 0:
        return 0
    return n + dummy_sum(n - 1)
````

Эта функция складывает все числа от 0 до n и возвращает сумму.

````python
dummy_sum(4)
````

```python
dummy_sum(4) -> 4 + dummy_sum(3)
dummy_sum(3) -> 3 + dummy_sum(2)
dummy_sum(2) -> 2 + dummy_sum(1)
dummy_sum(1) -> 1 + dummy_sum(0)
dummy_sum(0) -> 0
```

```python
dummy_sum(4) -> 4 + 6 = 10
dummy_sum(3) -> 3 + 3 = 6
dummy_sum(2) -> 2 + 1 = 3
dummy_sum(1) -> 1 + 0 = 1
dummy_sum(0) -> 0
```

ВАЖНОЕ ПРАВИЛО: если у вас рекурсивная функция, то должен быть хотя бы один `return`, который не использует результат выполнения функции, в которой мы находимся.


````python
def inf_dummy_sum(n):
    # if n == 0:
    #     return 0
    return n + inf_dummy_sum(n - 1)
````

```python
dummy_sum(4) -> 4 + dummy_sum(3)
dummy_sum(3) -> 3 + dummy_sum(2)
dummy_sum(2) -> 2 + dummy_sum(1)
dummy_sum(1) -> 1 + dummy_sum(0)
dummy_sum(0) -> 0 + dummy_sum(-1)
dummy_sum(-1) -> -1 + dummy_sum(-2)
...

```

````python
inf_dummy_sum(4)
````

## Необязательные аргументы с дефолтными значениями

````python
def my_fancy_function_with_punc(sentence, punc=',.;:!?"\''):
    for char in punc:
        sentence = sentence.replace(char, '')

    words = sentence.split()

    return words
````

````python
def my_fancy_function(sentence, additional_punc=''):
    punc = ',.;:!?"\'' + additional_punc
    for char in punc:
        sentence = sentence.replace(char, '')

    words = sentence.split()

    return words
````
