```python
text = "У меня есть 2 кота и 15 собак."

def double_number(match):
    number = int(match.group())
    return str(number * 2)

new_text = re.sub(r"\d+", double_number, text)

print(new_text)
```

```python
text = "Дата: 14.05.2026"

match = re.search(r"(\d{2})\.(\d{2})\.(\d{4})", text)

print(match.group())
print(match.group(0))
print(match.group(1))
print(match.group(2))
print(match.group(3))
```

```python
text = "Дата: 14.05.2026"

match = re.search(
    r"(?P<day>\d{2})\.(?P<month>\d{2})\.(?P<year>\d{4})",
    text
)

print(match.group("day"))
print(match.group("month"))
print(match.group("year"))
```

```python
print(match.groupdict())
```

```python
date_pattern = re.compile(
    r"(?P<day>\d{2})\.(?P<month>\d{2})\.(?P<year>\d{4})"
)

texts = [
    "Дата рождения: 01.09.2000",
    "Дата экспедиции: 14.05.2026",
    "Здесь даты нет"
]

for text in texts:
    match = date_pattern.search(text)
    if match:
        print(match.groupdict())
```
